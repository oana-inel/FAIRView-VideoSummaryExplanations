```python
dataset["Answer.explanation_type"] = ""
dataset["Answer.explanation_reason"] = ""


for i in range(len(dataset)):
    explanations = []
    reasons = []
    if str(dataset["Answer.representativeness.radio1"].iloc[i]).lower() == "true":
        explanations.append("explanation1")
    if str(dataset["Answer.representativeness.radio2"].iloc[i]).lower() == "true":
        explanations.append("explanation2")
    if str(dataset["Answer.representativeness.radio3"].iloc[i]).lower() == "true":
        explanations.append("explanation3")
    if str(dataset["Answer.representativeness.radio4"].iloc[i]).lower() == "true":
        explanations.append("explanation4")
        
    if str(dataset["Answer.motivation.reason1"].iloc[i]).lower() == "true":
        reasons.append("motivation1")
    if str(dataset["Answer.motivation.reason2"].iloc[i]).lower() == "true":
        reasons.append("motivation2")
    if str(dataset["Answer.motivation.reason3"].iloc[i]).lower() == "true":
        reasons.append("motivation3")
    if str(dataset["Answer.motivation.reason4"].iloc[i]).lower() == "true":
        reasons.append("motivation4")
    if str(dataset["Answer.motivation.reason5"].iloc[i]).lower() == "true":
        reasons.append("motivation5")
    if str(dataset["Answer.motivation.reason6"].iloc[i]).lower() == "true":
        reasons.append("motivation6")
    if str(dataset["Answer.motivation.reason7"].iloc[i]).lower() == "true":
        reasons.append("motivation7")
    if str(dataset["Answer.motivation.reason8"].iloc[i]).lower() == "true":
        reasons.append("motivation8")
    if str(dataset["Answer.other"].iloc[i]).lower() == "true":
        reasons.append("other")
    
    dataset["Answer.explanation_type"].iloc[i] = (",").join(explanations)
    dataset["Answer.explanation_reason"].iloc[i] = (",").join(reasons)
```

```python
dataset.to_csv("results_fairview_explanation_types_processed.csv", index=False)
```

```python
# Run CT metrics on the four explanations
import crowdtruth
from crowdtruth.configuration import DefaultConfig

class TestConfig(DefaultConfig):
    inputColumns = ["Input.videoStorageLink"]
    outputColumns = ["Answer.explanation_type"]
    
    # processing of a closed task
    open_ended_task = False
    
    def processJudgments(self, judgments):
        # pre-process output to match the values in annotation_vector
        return judgments
    
data, config = crowdtruth.load(
    data_frame = dataset,
    config = TestConfig()
)

#print(data['judgments'].head())

results = crowdtruth.run(data, config)

#print(results["units"].head())
results["annotations"].head()
```
