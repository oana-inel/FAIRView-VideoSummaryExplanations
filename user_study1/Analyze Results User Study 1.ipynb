```python
import pandas as pd
from IPython.display import display
pd.set_option('display.max_columns', 500)

dataset = pd.read_csv("crowd_results_userstudy1.csv")
dataset.head(5)
```

```python
# print total number of crowd workers in the task
print("Total number of crowd workers:", len(list(set(list(dataset["WorkerId"])))))
```

```python
# print the total amount of judgments per crowd worker
dataset.groupby(['WorkerId']).count().sort_values(['HITId'])
```

```python
# Average number of judgments per crowd workers
print("Average number of judgments per crowd workers:", dataset.groupby(['WorkerId']).count().mean()["HITId"])
```

```python
# identify the crowd workers answers for the explanation type and explanation reason

dataset["Answer.explanation_type"] = ""
dataset["Answer.explanation_reason"] = ""


for i in range(len(dataset)):
    explanations = []
    reasons = []
    if str(dataset["Answer.representativeness.radio1"].iloc[i]).lower() == "true":
        explanations.append("summary_wordcloud")
    if str(dataset["Answer.representativeness.radio2"].iloc[i]).lower() == "true":
        explanations.append("overlap_fraction")
    if str(dataset["Answer.representativeness.radio3"].iloc[i]).lower() == "true":
        explanations.append("overlap_wordcloud")
    if str(dataset["Answer.representativeness.radio4"].iloc[i]).lower() == "true":
        explanations.append("combined_wordcloud_and_fraction")
        
    if str(dataset["Answer.motivation.reason1"].iloc[i]).lower() == "true":
        reasons.append("motivation1")
    if str(dataset["Answer.motivation.reason2"].iloc[i]).lower() == "true":
        reasons.append("motivation2")
    if str(dataset["Answer.motivation.reason3"].iloc[i]).lower() == "true":
        reasons.append("motivation3")
    if str(dataset["Answer.motivation.reason4"].iloc[i]).lower() == "true":
        reasons.append("motivation4")
    if str(dataset["Answer.motivation.reason5"].iloc[i]).lower() == "true":
        reasons.append("motivation5")
    if str(dataset["Answer.motivation.reason6"].iloc[i]).lower() == "true":
        reasons.append("motivation6")
    if str(dataset["Answer.motivation.reason7"].iloc[i]).lower() == "true":
        reasons.append("motivation7")
    if str(dataset["Answer.motivation.reason8"].iloc[i]).lower() == "true":
        reasons.append("motivation8")
    if str(dataset["Answer.other"].iloc[i]).lower() != "nan":
        reasons.append("other")
    
    dataset["Answer.explanation_type"].iloc[i] = (",").join(explanations)
    dataset["Answer.explanation_reason"].iloc[i] = (",").join(reasons)
```

# Identify the number of votes for each visual explanation and motivation

```python
dataset["explanation_types"] = dataset["Answer.explanation_type"].str.split(",")
dataset["motivations"] = dataset["Answer.explanation_reason"].str.split(",")

def countList(lst, x): 
    count = 0
    for i in range(len(lst)): 
        if x in lst[i]: 
            count+= 1
          
    return count

explanations = ["combined_wordcloud_and_fraction", "overlap_fraction", "overlap_wordcloud", "summary_wordcloud"]
crowd_choices = list(dataset["explanation_types"])

print("Counts of representative visual explanations")
for explanation in explanations:
    print(explanation, countList(crowd_choices, explanation)) 
print("\n")    

motivations = ["motivation1", "motivation2", "motivation3", "motivation4", "motivation5", "motivation6", "motivation7"
             "motivation8", "other"]

# The list of motivations refer to the following:
# motivation1: It was useful to see the concepts of the video summary
# motivation2: It was useful to see which concepts from the original video were covered in the summary
# motivation3: It was useful to see the fraction of concepts in the video summary
# motivation4: It was useful to see the prominent concepts of the original video
# motivation5: It was useful to see the prominent concepts of the video summary
# motivation6: It was useful to see thee differences between summary and original video
# motivation7: It was useful to see the similarities between summary and original video
# motivation8: It was useful to see the concepts from the original video not covered by the summary
# other

crowd_choices = list(dataset["motivations"])

print("Counts of motivations")
for motivation in motivations:
    print(motivation, countList(crowd_choices, motivation))
    
```

# H0: The baseline explanation, Summary WordCloud, has the least utility for understanding the representativeness of video summaries.

```python
import numpy as np
table = np.zeros((400, 2))

for i in range(len(dataset)):
    if str(dataset["Answer.representativeness.radio1"].iloc[i]).lower() == "true":
        table[i,0] = 1
    if str(dataset["Answer.representativeness.radio2"].iloc[i]).lower() == "true":
        table[i,1] = 1

from statsmodels.sandbox.stats.runs import cochrans_q 

st, pvalue = cochrans_q(table)

print("Cochran's Q:", st)
print("p-value:", pvalue)
```
