```python
dataset["Answer.representative"] = "" 

for i in range(len(dataset)):
    if str(dataset["Answer.representativeness.same"].iloc[i]).lower() == "true":
        dataset["Answer.representative"].iloc[i] = "same"
    if str(dataset["Answer.representativeness.none"].iloc[i]).lower() == "true":
        dataset["Answer.representative"].iloc[i] = "none"
        
    if "__aq" in dataset["Input.explanation11"].iloc[i]:
        if str(dataset["Answer.representativeness.summary1"].iloc[i]).lower() == "true":
            dataset["Answer.representative"].iloc[i] = "summary2"
        if str(dataset["Answer.representativeness.summary2"].iloc[i]).lower() == "true":
            dataset["Answer.representative"].iloc[i] = "summary1"
    else:
        if str(dataset["Answer.representativeness.summary1"].iloc[i]).lower() == "true":
            dataset["Answer.representative"].iloc[i] = "summary1"
        if str(dataset["Answer.representativeness.summary2"].iloc[i]).lower() == "true":
            dataset["Answer.representative"].iloc[i] = "summary2"
    
```
