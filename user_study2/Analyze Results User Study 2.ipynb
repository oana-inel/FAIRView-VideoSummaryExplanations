```python
import pandas as pd
```

```python
from IPython.display import display
pd.set_option('display.max_columns', 500)

dataset = pd.read_csv("crowd_results_userstudy2_copy.csv")
dataset.head(5)
```

```python
print("Total number of crowd workers that participated in the study:")
print(len(list(set(list(dataset["WorkerId"])))))
```

```python
print("Total number of judgments per crowd worker:")
dataset.groupby(['WorkerId']).count().sort_values(['HITId'])
```

```python
print("Average number of judgments per crowd worker:")
dataset.groupby(['WorkerId']).count().mean()['HITId']
```

```python
dataset["Answer.representative"] = "" 

for i in range(len(dataset)):
    if str(dataset["Answer.representativeness.same"].iloc[i]).lower() == "true":
        dataset["Answer.representative"].iloc[i] = "same"
    if str(dataset["Answer.representativeness.none"].iloc[i]).lower() == "true":
        dataset["Answer.representative"].iloc[i] = "none"
        
    if "__aq" in dataset["Input.explanationSummary1"].iloc[i]:
        if str(dataset["Answer.representativeness.summary1"].iloc[i]).lower() == "true":
            dataset["Answer.representative"].iloc[i] = "summary2"
        if str(dataset["Answer.representativeness.summary2"].iloc[i]).lower() == "true":
            dataset["Answer.representative"].iloc[i] = "summary1"
    else:
        if str(dataset["Answer.representativeness.summary1"].iloc[i]).lower() == "true":
            dataset["Answer.representative"].iloc[i] = "summary1"
        if str(dataset["Answer.representativeness.summary2"].iloc[i]).lower() == "true":
            dataset["Answer.representative"].iloc[i] = "summary2"
    
```
